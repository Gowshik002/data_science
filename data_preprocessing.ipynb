```python
import numpy as np
import pandas as pd
```

```python
data=pd.Series([0,1,2,3,4,5,np.nan,6,7,8])
```

```python
data
```

#to detect null **values**

```python
data.isnull() 
```

#  detect existing values (non-missing )

#creating a dataframe

```python
dict1={
    0:[1,4,7,np.nan],
    1:[2,5,np.nan,np.nan],
    2:[3,np.nan,np.nan,np.nan],
    3:[np.nan,np.nan,np.nan,np.nan]
}
```

```python
df=pd.DataFrame(dict1)
```

```python
df
```

#drop a entire row if one nan value present

```python
df.dropna()
```

#deletes the row in which all the values are null

```python
df.dropna(how='all')
```

#deletes the columns in which all the values are null

```python
df.dropna(axis=1,how='all')
```

#threshold requires many non-nan values

```python
df.dropna(thresh=1)
```

```python
df.dropna(axis=1,thresh=1)
```

#filling the null values

   1.syntax-->df.fillna(val)

   2.returns a new DataFrame


```python
df_fill=df.fillna(10)
```

```python
df_fill
```

```python
df_fill=df.fillna({0:10,1:20,2:30,3:40})
```

```python
df_fill
```

```python
df_fill=df.fillna(method='ffill')
```

```python
df_fill
```

```python
df_fill=df.fillna(method='ffill',limit=2)
df_fill
```

#df.mean()


*   gives the mean
*   to fill the missing values by the mean values we use mean()





```python
df_mean=df.fillna(df.mean())
df_mean
```

#.replace(old_val,new_val)


*   replace the old values with new values
*   values can be anything




```python
ser1=pd.Series([1,2,3,100,45,100,9,8,5])
ser1
```

```python
ser1.replace(100,1000)      #to replace 100 with 1000
```

```python
ser1.replace([100,1000,2],500)        #to replace multiple values with single value
```

```python
ser1.replace({100:np.nan,2:250})
```
