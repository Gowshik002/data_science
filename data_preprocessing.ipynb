```python
import pandas as pd
import numpy as np
```

```python
data=pd.Series([1,2,3,4,5,np.nan,6,7,8])
data
```

```python
data.isnull()
```

```python
data.notnull()
```

```python
dict1={
    0:[1,4,7,np.nan],
    1:[2,5,np.nan,np.nan],
    2:[3,np.nan,np.nan,np.nan],
    3:[np.nan,np.nan,np.nan,np.nan]
}
dict1
```

```python
df=pd.DataFrame(dict1)
df              
```

```python
df.dropna()             #drops entire row if one nan value is present
```

```python
df.dropna(how='all')       #deletes the row in which all the values are null
```

```python
df.dropna(axis=1,how='all')     #deletes the column in which all the values are null
```

```python
df
```

```python
df.dropna(thresh=1)     #threshold requires that many 'non nan' values
```

#filling the null values


*   syntax->df.fillna(val)
*   returns a new DataFrame



```python
df
```

```python
df_fill=df.fillna(10)
df_fill
```

```python
df_fill=df.fillna({0:10,1:20,2:30,3:40})          #to fill particular values in particular columns
df_fill
```

```python
df_fill=df.fillna(method='ffill')    #fills the nan values by previous observation/row value
df_fill
```

```python
df_fill=df.fillna(method='ffill',limit=2)        #if you specify the limit then for those many values in the column it will fill the values
df_fill
```

```python
df_fill=df.fillna(method='ffill',limit=2,axis=1)
df_fill
```

#df mean()


*    gives the mean
*   to fill the missing values by the mean values we use mean()



```python
df_mean=df.fillna(df.mean())
df_mean
```

#.replace(old_val,new_val)

*   replace old values with new values
*   values can be anything

```python
ser1=pd.Series([1,2,100,4,5,100,7,8,100])
ser1
```

```python
ser1.replace(100,1)        #to replace 100 with 1
```

```python
ser1.replace([100,1,2],50)           #to replace multiple values with single value
```

```python
ser1.replace({100:np.nan,1:50})
```
